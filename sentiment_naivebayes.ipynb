```python
import torch
from datasets import load_dataset,Dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report
```

```python
HUGGINGFACE_PATH = "uitnlp/vietnamese_students_feedback"
NUM_CORE = 2 # Colab Maximum Core

dataset = load_dataset(
    path=HUGGINGFACE_PATH,
    num_proc=NUM_CORE,
    split="train",
    verification_mode="no_checks",
    cache_dir="cache")
```

```python
data_df=dataset.to_pandas()
data_df
```

```python
x = data_df["sentence"]
y = data_df["sentiment"]

```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
vectorizer=CountVectorizer()
x_train_vec=vectorizer.fit_transform(x_train)
x_test_vec=vectorizer.transform(x_test)
modal=MultinomialNB()
modal.fit(x_train_vec,y_train)
```

```python
y_pred=modal.predict(x_test_vec)
accuracy=accuracy_score(y_test,y_pred)
print("accuracy",accuracy)
Report=classification_report(y_test,y_pred)
print(Report)
```

```python
label = {0: "NEGATIVE", 1: "NEUTRAL", 2: "POSITIVE"}
new_review=["Thầy giáo giảng buồn ngủ."]
new_review_vec=vectorizer.transform(new_review)
pred_sentiment=modal.predict(new_review_vec)
print("Predictful sentiment",label[pred_sentiment[0]])
```
